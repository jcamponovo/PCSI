```python
%matplotlib notebook
#En dynamique
from math import cos,pi,sin,atan2
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from ipywidgets import interact, widgets


t = [0.001*i for i in range(1001)]

def f1(A,f):
    global t
    return [A*cos(2*pi*f*val) for val in t]

def f2(B,f,phi):
    global t
    return [B*cos(2*pi*f*val + phi) for val in t]

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
fig.show()

@interact(A=(0,5,0.1),B=(0,5,0.1),f=(2,6,0.1),phi=(-pi,pi,pi/8))
def setx0(A=4.,B=3.,f=4.,phi=pi/2):
    x1=f1(A,f)
    x2=f2(B,f,phi)
    xsomme=[x1[i]+x2[i] for i in range(len(x1))]
    ax1.clear()
      
    ax1.spines['right'].set_color('none')
    ax1.spines['top'].set_color('none')
    ax1.spines['bottom'].set_position(('data',0))
    ax1.spines['left'].set_position(('data',0))
    ax1.xaxis.set_ticks_position('bottom')
    ax1.yaxis.set_ticks_position('left')
    ax1.set_xlabel("t (en s)")
    ax1.set_ylabel("x (en m)")
    ax1.set_xlim([0,1])
    ax1.set_ylim([-11,17])
    ax1.set_xticks([0,0.5,1])
    n = round(phi*8/pi)
    titre=""
    rq = "Amplitude diffÃ©rente de la somme A + B"
    if n==-8:
        dep = "-\pi"
        titre = "$s_1\ et\ s_2$ sont en opposition de phase"
        rq = "L'amplitude vaut la diffÃ©rence A - B"
    elif n==8:
        dep = "+\pi"
        titre = "$s_1\ et\ s_2$ sont en opposition de phase"
        rq = "L'amplitude vaut la diffÃ©rence A - B"
    elif n==0:
        dep =""
        titre= "$s_1\ et\ s_2$ sont en phase"
        rq = "L'amplitude vaut la somme A + B"
    elif n==-4:
        dep = "-\dfrac{\pi}{2}"
        titre = "$s_2$ est en quadrature retard par rapport Ã  $s_1$"
    elif n==4:
        dep = "+\dfrac{\pi}{2}"
        titre = "$s_2$ est en quadrature avance par rapport Ã  $s_1$"
    elif n==-2:
        dep = "-\dfrac{\pi}{4}"
    elif n==2:
        dep = "+\dfrac{\pi}{4}"
    elif n==-6:
        dep = "-\dfrac{3\pi}{4}"
    elif n==6:
        dep = "+\dfrac{3\pi}{4}"
    elif n==-1:
        dep = "-\dfrac{\pi}{8}"
    elif n==1:
        dep = "+\dfrac{\pi}{8}"      
    elif n<0:
        dep = "-\dfrac{"+str(-n)+"\pi}{8}"
    else:
        dep = "+\dfrac{"+str(n)+"\pi}{8}"   
    
    
    ax1.plot(t,x1,"blue",label="$s_1(t)\ =\ {}\ \cos(2\ \pi\ f\ t)$".format(A))
    ax1.plot(t,x2,"red",label="$s_2(t)\ =\ {}\ \cos(2\ \pi\ f\ t\ {})$".format(B,dep))
    ax1.plot(t,xsomme,"green",label="$A\cos(2\ \pi\ f\ t)+B\cos(2\ \pi\ f\ t\ +\phi)$\n"+rq)
    ax1.legend(loc="upper right",framealpha=1.)
    ax1.set_title(titre)
    
  
setx0()

    
```
